```python
df=pd.read_csv('/content/California_Houses.csv')
```

```python
# Print the first 5 rows of the data
df.head()
```

```python
df.info()
```

```python
sns.lineplot(data=df['Tot_Rooms'],label="Total Rooms")
sns.lineplot(data=df['Tot_Bedrooms'],label="Total Bedrooms")
plt.show()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(8,6))
v=df.query('Median_Income<6').Median_Income
sns.barplot(x=df['Median_Age'],y=v)
```

```python
plt.figure(figsize=(14,7))
sns.heatmap(data = df.corr(), annot =True)
```

```python
plt.figure(figsize=(14,7))
sns.scatterplot(x=df['Median_Income'], y=df['Median_House_Value'], hue=df['Median_Age'])
```

```python
plt.figure(figsize=(14,7))
cond_age = df.query('30 <= Median_Age <= 40').Median_Age
sns.kdeplot(data=df, x='Distance_to_LA', hue=cond_age, shade=True)
```

```python
import plotly.express as px
fig = px.scatter(df, x='Median_House_Value',y='Households', color="Population")
fig.show()
```
